```python
import scipy.fft
import numpy as np
import matplotlib.pyplot as plt

sample_rate = 128  # Hz
wdw_seconds = 2 # seconds
freq = 3  # Hz

a = np.arange(0, wdw_seconds * sample_rate)
b = np.sin((2 * np.pi) * freq * a / sample_rate)

_, ax1 = plt.subplots(1)
ax1.plot(b)

b_fft = scipy.fft.fft(b)
freq_axis = np.linspace(0, sample_rate, sample_rate * wdw_seconds)

_, ax2 = plt.subplots(1)
ax2.plot(freq_axis, np.abs(b_fft))
```
